```python
import numpy as np
import time
import torch
import torch.nn as nn
import os
#import visdom
import random
from tqdm import tqdm as tqdm
import pickle

from CS492_CrowdCounting.cannet import CANNet
from CS492_CrowdCounting.my_dataset import CrowdDataset


if not os.path.exists('./checkpoints'):
    os.mkdir('./checkpoints')

train_loss_list=[]
epoch_list=[]
test_error_list=[]

with open('/content/drive/My Drive/AI_Project2/checkpoints/train_loss_list_ver1.pkl', 'rb') as f:
  train_loss_list = pickle.load(f)
with open('/content/drive/My Drive/AI_Project2/checkpoints/test_error_list_ver1.pkl', 'rb') as f:
  test_error_list = pickle.load(f)
LAST_EPOCH  = len(train_loss_list)
LAST_EPOCH = LAST_EPOCH - 1 if len(train_loss_list)!=len(test_error_list) else LAST_EPOCH
train_loss_list = train_loss_list[:LAST_EPOCH+1]
test_error_list = test_error_list[:LAST_EPOCH+1]
```

```python
# epoch:85 error:75.2353589613359 min_mae:74.85619423415635 min_epoch:76
```

```python
import matplotlib.pyplot as plt
import numpy as np
test_error_list = test_error_list
x = np.arange(0, len(train_loss_list),1)
plt.rcParams["figure.figsize"] = (10, 10)
plt.title('9x9 pooling added', fontsize=30)
plt.plot(x,train_loss_list,'-r',label='train loss')
plt.plot(x,test_error_list,'-b', label = 'test loss')
plt.legend(loc='upper left')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.show()
```
